```python
json_dict = {}
error = []
for word in tqdm(data):
    a = 'https://oxford.mister5.net/?word='
    b = word.replace('\n','')
    url = a+b
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    find_pattern = re.compile(r'[(]')
    li_tag = soup.find_all('li', htag = 'li')
    li_list =[]
    try:
        for i in li_tag:
            span_tag = i.find_all('span', class_='cf')
            if span_tag != []:
                match_result = find_pattern.findall(str(i))
                if match_result == []:
                    cf_tag = i.find('span', class_='cf')
                    x_tag = i.find('span', class_='x')
                    chn_tag = i.find('chn')
                    li_list.append({'pattern': cf_tag.get_text(), 'example': x_tag, 'trans': chn_tag.get_text()})
        if li_list != []:
            json_dict[b] = li_list
            
    except:
        error.append(b)
```

```python
category = {}
for i in data:
    for j in data[i]:
        pattern = j['pattern'].split(' ',1)
        try:
            find_something = re.compile(i)
            match_result = find_something.findall(j['pattern'].lower())
            if match_result != []:
                if pattern[1] not in category.keys():
                    category[pattern[1]] = 1
                else:
                    category[pattern[1]] += 1
            if match_result == []:
                if j['pattern'] not in category.keys():
                    category[j['pattern']] = 1
                else:
                    category[j['pattern']] += 1
        except:
            print(j)
```

```python
cate_list = []
for i in category:
    cate_list.append((i,category[i]))
cate_sort = sorted(cate_list, key = lambda x : x[1], reverse = True)
```
